```python
import pandas as pd
df=pd.read_csv("BOLLYWOOD.csv")
df.head()
```

```python
#now starting doing eda

df.groupby('release_year')['profit'].agg(['sum','mean','max'])
```

```python
df.groupby('release_year')['profit'].agg(['max']).plot(kind='bar')
```

```python
df[['movie_name','profit']].sort_values(by='profit', ascending=False).head(10)
```

```python
df[['movie_name','profit']].sort_values(by='profit', ascending=False).head(10)
```

```python
df[['budget','profit']].corr()
```

```python
df.groupby('genre')['profit'].mean().sort_values(ascending=False)
```

```python
df.groupby('imdb_rating')['profit'].mean()
```

```python
df['release_year'].value_counts().sort_index()
```

```python
df['profit'].describe()
```

```python
import matplotlib.pyplot as plt
df['release_year'].value_counts().plot(kind='bar')
plt.title("Movies Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
```

```python
df.groupby('release_year')['profit'].sum().plot(kind='bar')
plt.title("Total Profit Per Year")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
```

```python
plt.scatter(df['budget'], df['profit'])
plt.xlabel("Budget")
plt.ylabel("Profit")
plt.title("Budget vs Profit")
plt.show()
```

```python
plt.hist(df['profit'], bins=20)
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit Distribution")
plt.show()
```

```python
df['result'] = df['profit'].apply(lambda x: 'Hit' if x>0 else 'Flop')
df['result'].value_counts()

df['result'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Hit vs Flop Movies")
plt.ylabel("")
plt.show()
```
